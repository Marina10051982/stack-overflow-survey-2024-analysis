# Аналіз даних Stack Overflow Developer Survey
У цьому ноутбуці проведено аналіз результатів опитування розробників Stack Overflow з використанням бібліотеки Pandas.
Мета роботи — дослідити структуру даних, популярність Python, досвід роботи, формат зайнятості та рівень компенсації.

## Імпорт бібліотек та завантаження даних

```python
import pandas as pd

df = pd.read_csv('survey_results_public.csv', low_memory=False)
schema = pd.read_csv('survey_results_schema.csv')
df.head()
```

## Завдання 1. Визначення кількості респондентів

```python
total_respondents = df.shape[0]
total_respondents
```

## Завдання 2. Кількість респондентів, які відповіли на всі питання

```python
questions = set(schema['qname'].dropna())
columns = set(df.columns)

valid_cols = list(questions.intersection(columns))

complete_respondents = df[valid_cols].notna().all(axis=1).sum()
complete_respondents
```

## Завдання 3. Статистичні показники досвіду роботи

```python
mean_exp = df['WorkExp'].mean().round(2)
median_exp = df['WorkExp'].median()
mode_exp = df['WorkExp'].mode()[0]

mean_exp, median_exp, mode_exp
```

## Завдання 4. Частка розробників, які працюють віддалено

```python
remote_workers = df['RemoteWork'].str.contains('Remote', case=False, na=False).sum()
remote_workers
```

## Завдання 5. Відсоток респондентів, які використовують Python

```python
python_users = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)

python_percentage = round(python_users.mean()*100,2)
python_percentage
```

## Завдання 6. Кількість респондентів, які навчались через онлайн‑курси

```python
online_learners = df['LearnCode'].str.contains('Online', case=False, na=False).sum()
online_learners
```

## Завдання 7. Середня та медіанна компенсація Python‑розробників по країнах

```python
python_devs = df[df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)]

salary_stats = (
    python_devs.groupby('Country')['ConvertedCompYearly']
    .mean()
    .round(0)
    .astype('Int64')
    .sort_values(ascending=False)
)

salary_stats.head(10)
```

## Завдання 8. Рівень освіти 5 найбільш оплачуваних респондентів

```python
top5 = (
    df[['ConvertedCompYearly','EdLevel']]
    .dropna()
    .sort_values('ConvertedCompYearly', ascending=False)
    .head(5)
)

top5['ConvertedCompYearly'] = top5['ConvertedCompYearly'].astype(int)

top5
```

## Завдання 9. Аналіз популярності Python по віковим категоріям (необов'язкове)

```python
python_users = df[df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)]

python_by_age = python_users.groupby('Age').size()
total_by_age = df.groupby('Age').size()

python_percent_age = ((python_by_age / total_by_age) * 100).round(2)

python_percent_age.reset_index(name='Python_%')
```

## Завдання 10. Індустрії серед високооплачуваних віддалених працівників (необов'язкове)

```python
q75 = df['ConvertedCompYearly'].quantile(0.75)

target = df[
    (df['RemoteWork'] == 'Remote') &
    (df['ConvertedCompYearly'] >= q75)
]

top_industries = (
    target['Industry']
    .dropna()
    .value_counts()
    .reset_index()
)

top_industries.columns = ['Industry','Count']
top_industries
```

## Висновок
У роботі проведено аналіз результатів опитування Stack Overflow Developer Survey. Було визначено кількість респондентів, проаналізовано досвід роботи, популярність Python, формат роботи та рівень компенсації. Отримані результати демонструють актуальні тенденції на ринку праці розробників.
